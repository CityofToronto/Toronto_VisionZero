```
# Features for the classification
features = list(df4.columns[2:7]) + list(df4.columns[11:])  # NOT including: 'collision_no', 'year', '# involved', '# injured', '# KSI', '# fatalities'
                                                                             
features
```

```
df4_n19 = df4[df4['year']!=2019]  # all years except 2019!!!

x = df4_n19[features]   # NO y
```

```
# Standardizing the features
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x = sc.fit_transform(x)
```

```
from sklearn.cluster import KMeans

WCSS = []

for k in range(1, 100):
    km = KMeans(n_clusters=k, init='k-means++', max_iter=300, n_init=10, random_state=0)  # | KMeans(n_clusters=k) 
                                                                                # 'k-means++' ensures that you get don’t fall into the random initialization trap.
    km = km.fit(x)
    WCSS.append(km.inertia_)
```

We graph the relationship between the number of clusters and Within Cluster Sum of Squares (**WCSS**) then we select the number of clusters where the change in WCSS begins to level off (elbow method). <br>
WCSS is defined as the **sum of the squared distances** between each member of the cluster and its centroid. <br>
As k increases, WCSS tends to zero. Imagine we set k to its maximum value n (where n is number of samples) each sample will form its own cluster meaning sum of squared distances equals zero.
